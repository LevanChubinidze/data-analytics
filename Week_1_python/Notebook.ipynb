```python
#df[df["Category"]== 'Electronics'] #selecting rows where category is Electronics
df[(df["Category"]== 'Electronics') & (df["Price"] > 1000)] #selecting rows where category is Electronics and price is greater than 1000
```

```python
df[(df["Category"]== 'Electronics') | (df["Price"] > 1000)] #selecting rows where category is Electronics or price is greater than 1000
```

```python
#base filtering
df[df['Quantity'] > 3] #selecting rows where quantity is greater than 3
```

```python
#Advance filtering
#df[df["CustomerName"].str.endswith("a")]
df[df["Country"].isin(["USA", "Spain"])] #selecting rows where country is USA or Spain
```

```python
#Updating Data
df.loc[df["CustomerName"] == "Mason Lewis", "Product"] = "Smartphone" #updating product for Mason Lewis
df[df["CustomerName"] == "Mason Lewis"] #checking if the update was successful  
df.loc[df["Country"]== "USA", "Country"] = "United States" #updating country name from USA to United States
df[df["Country"]== "United States"] #checking if the update was successful
```

```python
df["Country"] = df["Country"].str.upper() #converting country names to uppercase
#df["Country"] #checking if the update was successful
```

```python
df = df.drop(3) #dropping row with index 3
#df = df.drop(columns=["Price"]) #dropping Price column
```

```python
df.dropna() #dropping rows with missing values
df.fillna({"OrderID": 0}, inplace=True) #filling missing values in OrderID column with 0
```

```python
#Rename columns
df.rename(columns={"OrderID": "Order ID"}, inplace=True) #renaming OrderID column
```

```python
df.groupby("Country")["Price"].sum() #calculating total price for each category
df.sort_values("Price", ascending=False) #sorting data by price in descending order
```

```python
df.to_csv("Updated_Order.csv", index=False) #saving updated data to a new CSV file
```
