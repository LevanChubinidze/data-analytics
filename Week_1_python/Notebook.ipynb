```python
    #selecting rows where category is Electronics
#df[df["Category"]== 'Electronics']
    #selecting rows where category is Electronics and price is greater than 1000
df[(df["Category"]== 'Electronics') & (df["Price"] > 1000)] 
```

```python
    #selecting rows where category is Electronics or price is greater than 1000
df[(df["Category"]== 'Electronics') | (df["Price"] > 1000)] 
```

```python
#base filtering
    #selecting rows where quantity is greater than 3
df[df['Quantity'] > 3] 
```

```python
    #Advance filtering
#df[df["CustomerName"].str.endswith("a")]
    #selecting rows where country is USA or Spain
df[df["Country"].isin(["USA", "Spain"])] 
```

```python
#Updating Data
    #updating product for Mason Lewis
df.loc[df["CustomerName"] == "Mason Lewis", "Product"] = "Smartphone"
    #checking if the update was successful
df[df["CustomerName"] == "Mason Lewis"]
    #updating country name from USA to United States
df.loc[df["Country"]== "USA", "Country"] = "United States"
    #checking if the update was successful
df[df["Country"]== "United States"] 
```

```python
    #converting country names to uppercase
df["Country"] = df["Country"].str.upper()
    #checking if the update was successful
#df["Country"]
```

```python
    #dropping row with index 3
df = df.drop(3)
    #dropping Price column
#df = df.drop(columns=["Price"])
```

```python
    #dropping rows with missing values
df.dropna()
    #filling missing values in OrderID column with 0
df.fillna({"OrderID": 0}, inplace=True) 
```

```python
#Rename columns
    #renaming OrderID column to Order ID
df.rename(columns={"OrderID": "Order ID"}, inplace=True) 
```

```python
    #calculating total price for each category
df.groupby("Country")["Price"].sum() 
    #sorting data by price in descending order
df.sort_values("Price", ascending=False) 
```

```python
    #saving updated data to a new CSV file
df.to_csv("Updated_Order.csv", index=False)
```
